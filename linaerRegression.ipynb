```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "height":[169,180,179,156,134],
    "weight":[50,60,55,64,58]
}
```

```python
df = pd.DataFrame(data)
x = df[["height"]]
y = df["weight"]
model = LinearRegression().fit(x,y)
```

```python
print("the intercept: ", model.intercept_)
print("the slope", model.coef_[0])
```

```python
height = 189
predicted_weight= model.predict([[height]])[0]
print(f"predicted weight for height {height} cm: {predicted_weight:.2f} kg")
```
