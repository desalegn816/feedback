```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor









```

```python
# Dataset
X = np.array([
    [1,6],
    [2,4],
    [3,7],
    [6,8],
    [7,1],
    [8,4]
])

y = np.array([7,8,16,44,50,68])
```

```python
# Query point
Q = np.array([[4,2]])

print("Query Point:", Q)
```

```python
# 1-NN
knn1 = KNeighborsRegressor(n_neighbors=1, metric='euclidean')
knn1.fit(X,y)
pred1 = knn1.predict(Q)

print("\nEuclidean Distance")
print("1-NN Prediction:", pred1)
```

```python
# 3-NN
knn3 = KNeighborsRegressor(n_neighbors=3, metric='euclidean')
knn3.fit(X,y)
pred3 = knn3.predict(Q)

print("3-NN Prediction:", pred3)
```

```python
# -----------------------------
# Manhattan Distance
# -----------------------------

# 1-NN
knn1_m = KNeighborsRegressor(n_neighbors=1, metric='manhattan')
knn1_m.fit(X,y)
pred1_m = knn1_m.predict(Q)

print("\nManhattan Distance")
print("1-NN Prediction:", pred1_m)
```

```python
# 3-NN
knn3_m = KNeighborsRegressor(n_neighbors=3, metric='manhattan')
knn3_m.fit(X,y)
pred3_m = knn3_m.predict(Q)

print("3-NN Prediction:", pred3_m)
```
